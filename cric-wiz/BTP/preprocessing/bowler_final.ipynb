```python
import pandas as py
import numpy as np
df = py.read_csv(r'D:\ML\BTP\preprocessing\bowlerMatchByMatchData.csv')
```

```python
df
```

```python
df=df[['bowler','venue','batting_team','dream 11 score']]
```

```python
bowlerData = py.read_csv(r'D:\ML\BTP\preprocessing\bowlerData.csv')
```

```python
bowlerData = bowlerData[['bowler','Matches_Played','avg_drm11_score', 'Economy', 'strike_rate']]
```

```python
bowlerOppositon = py.read_csv(
    r'D:\ML\BTP\preprocessing\bowlerOppositionWiseData.csv')
```

```python
bowlerOppositon = bowlerOppositon[[
    'bowler', 'batting_team', 'bowlerOppositionWiseMatches_Played', 'bowlerOppositionWiseavg_drm11_score', 'bowlerOppositionWiseEconomy', 'bowlerOppositionWisestrike_rate']]
```

```python
bowlerVenue = py.read_csv(r'D:\ML\BTP\preprocessing\bowlerVenueWiseData.csv')
```

```python
bowlerVenue = bowlerVenue[[
    'bowler', 'venue', 'bowlerVenueMatches_Played', 'bowlerVenueavg_drm11_score', 'bowlerVenueEconomy', 'bowlerVenuestrike_rate']]
```

```python
venuedata = py.read_csv(
    r'D:\ML\BTP\preprocessing\VenueData.csv')
venuedata = venuedata[['venue', 'Venue_Matches_Played',
                      'Venue_Avg_Batsmen_Strike_Rate', 'Venue_Avg_Batsmen_Average', 'Venue_Avg_Bowler_Economy', 'Venue_Avg_Bowler_Average', 'Venue_dream11']]
```

```python

df = df.merge(bowlerData, on='bowler', how='left')
df = df.merge(bowlerOppositon, on=[
    'bowler', 'batting_team'], how='left')
df = df.merge(bowlerVenue, on=['bowler', 'venue'], how='left')
df = df.merge(venuedata, on=['venue'], how='left')
# df.drop('match_id', axis=1, inplace=True)
# df.to_csv('final_bowler.csv', index=False)
```

```python
df.isnull().sum()
```

```python
df['Venue_Matches_Played'].fillna(
    value=df['Venue_Matches_Played'].mean(), inplace=True)
df['Venue_Avg_Batsmen_Strike_Rate'].fillna(
    value=df['Venue_Avg_Batsmen_Strike_Rate'].mean(), inplace=True)
df['Venue_Avg_Batsmen_Average'].fillna(
    value=df['Venue_Avg_Batsmen_Average'].mean(), inplace=True)
df['Venue_Avg_Bowler_Economy'].fillna(
    value=df['Venue_Avg_Bowler_Economy'].mean(), inplace=True)
df['Venue_Avg_Bowler_Average'].fillna(
    value=df['Venue_Avg_Bowler_Average'].mean(), inplace=True)
df['Venue_dream11'].fillna(
    value=df['Venue_dream11'].mean(), inplace=True)
# df.to_csv('final_bowler.csv', index=False)
```

```python
df.to_csv('final_bowler.csv', index=False)
```
